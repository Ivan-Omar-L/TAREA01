9)Escribir una funcion que reciba un numero entero positivo y devuelva su factorial

```
def factorial(entero):
  i = 1
  fact = 1 
  while i < entero :
    i = i + 1
    fact = fact * i 
    res = fact

  print("resultado del factorial =", res)

a =int (input("ingrese un numero:"))
factorial(a)
print("fin del programa")
```
